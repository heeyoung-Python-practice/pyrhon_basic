```python
lst = []

for n in range(1,11):
    lst.append(n ** 2)

# 리스트 내포
lst = [n ** 2 for n in range(1,11)]
print(lst)
```

```python
lst = [i ** 2 for i in range(1,11) if i % 2] # 홀수만
print(lst)
```
