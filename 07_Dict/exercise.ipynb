```python
# 1.shallow copy
a = {'a': 1, 'b': 2, 'c': 3}
b = a # 사전 레퍼런스의 복사 (공유됨)
print(a)
print(b)

a['a'] = 100
print(a)
print(b)
```

```python
# 2.deep copy
c = {'a': 1, 'b': 2, 'c': 3}
d = c.copy()
print(c)
print(d)
```

```python

c['a'] = 300
print(c)
print(d) # 변경되지 않음.
```

```python
c={'a':[1,2,3]}
d = c.copy()

c['a'][0] = 999

print(c)
print(d)
```

```python
a = {'a': [1,2,3], 'b':100}
b = a.copy()
print(a) #{'a': [1, 2, 3], 'b': 100}
print(b) #{'a': [1, 2, 3], 'b': 100}
```

```python
a['b'] = 200
a['a'][0] = 999
print(a) #{'a': [999, 2, 3], 'b': 200}
print(b) #{'a': [999, 2, 3], 'b': 100}
```

```python
# value에 dict가 있다면
a = {'user':{'name':'김동현','age':20},'amount':123}
b = a.copy();
print(a)
print(b)
```

```python
a['user']['name'] = '김동팔'
print(a)
print(b)
```

```python
import copy

original = {
    'name':'capybara',
    'score' : [100,90,80]
}

deep = copy.deepcopy(original)

original['score'][0] = 999

print(original)
print(deep)
```

```python
# update 메소드를 통한 두 dict의 병합
# 동일한 키값이 있다면, 인수로 주어진 dict의 아이템으로 덮어쓰기 된다.
dic1 = {'a':10, 'b':20}
dic2 = {'c':100, 'd':300}
dic1.update(dic2)

print(dic1) #{'a': 10, 'b': 20, 'c': 100, 'd': 300}
print(dic2) #{'c': 100, 'd': 300}
```

```python
# dict의 정렬 : sorted
numbers = {'first': 1, 'second': 2, 'third': 3, 'Fourth': 4}
numbers['hundred'] = 100
numbers['fifth'] = 5

print(list(numbers)) #['first', 'second', 'third', 'Fourth', 'hundred', 'fifth']

```

```python
# key값 정렬
print(numbers)
print(sorted(numbers)) #['Fourth', 'fifth', 'first', 'hundred', 'second', 'third']
```

```python
# value값 정렬
print(numbers.values())
print(sorted(numbers.values())) #[1, 2, 3, 4, 5, 100]
```

```python
# key정렬 (value값 정렬에 따라)
print(sorted(numbers, key=numbers.__getitem__))
```
