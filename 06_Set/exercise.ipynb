```python
ohgiraffers = set(["Monkey", "Tiger", "Wolf"])
print(ohgiraffers)    # {'Tiger', 'Monkey', 'Wolf'}

ohgiraffers.update(["Monkey", "Wolf", "Tiger", "Squirrel", "Monkey"])
print(ohgiraffers)    # {'Tiger', 'Squirrel', 'Monkey', 'Wolf'}
```

```python
ohgiraffers = {'Pig', 'Squirrel', 'Bear', 'Gorilla'}

ohgiraffers.remove('Elephant')
print(ohgiraffers)    #  {'Bear', 'Squirrel', 'Gorilla'}
```

```python
ohgiraffers.discard('Elephant')
print(ohgiraffers)    #  {'Bear', 'Squirrel', 'Gorilla'}
```

```python
ohgiraffers = {'Pig', 'Squirrel', 'Bear', 'Gorilla'}

ohgiraffers.pop()
print(ohgiraffers)    # {'Squirrel', 'Gorilla'}
```

```python
javaTeam = {'Squirrel','Gorilla', 'Tiger', 'Sheep', 'monkey', 'wolf'}
pythonTeam = {'Pig', 'Squirrel', 'Bear', 'Gorilla'}

ohgiraffers = javaTeam.union(pythonTeam)
print(ohgiraffers)              # {'wolf', 'Sheep', 'Tiger', 'Gorilla', 'monkey', 'Squirrel', 'Pig', 'Bear'}
```

```python
new_ohgiraffers = javaTeam | pythonTeam
print(new_ohgiraffers)
```

```python
print(javaTeam.intersection(pythonTeam))
print(javaTeam & pythonTeam)
```

```python
print(javaTeam.difference(pythonTeam))
print(javaTeam - pythonTeam)
```

```python
print(javaTeam.symmetric_difference(pythonTeam))
print(javaTeam ^ pythonTeam)
```
